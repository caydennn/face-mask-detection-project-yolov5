# Create a Data Config file

We need create a **mask-data.yaml** and place it inside yolov5/data folder. 
This config file makes it convenient for us to pass it as an argument during training to inform yolov5 where to find our training data, the number of classes and the label indexes.



Since we have two classes (no_mask and mask), we defined num_classes to be 2.

````python
num_classes = 2
````

The contents of **mask-data.yaml** should be as follows:

```
train: ../face_mask_dataset/images/train
val: ../face_mask_dataset/images/valid
test: ../face_mask_dataset/images/test

nc: 2
names: ['no_mask', 'mask']

```

Run the cells below to create mask-data.yaml

````python
#customize iPython writefile so we can write variables
from IPython.core.magic import register_line_cell_magic

@register_line_cell_magic
def writetemplate(line, cell):
    with open(line, 'w') as f:
        f.write(cell.format(**globals()))
````

````python
%%writetemplate yolov5/data/mask-data.yaml

train: ../face_mask_dataset/images/train
val: ../face_mask_dataset/images/valid
test: ../face_mask_dataset/images/test

nc: {num_classes}
names: ['no_mask', 'mask']

````

````python
%cat yolov5/data/mask-data.yaml
````

# Define model configuration and architecture

The original model configuration of yolov5s.pt defines 80 classes. 
We will modify the number of classes to fit our use case (which is 2 number of classes)

````python
%%writetemplate yolov5/models/custom_yolov5s.yaml

# parameters
nc: {num_classes}  # number of classes
depth_multiple: 0.33  # model depth multiple
width_multiple: 0.50  # layer channel multiple

# anchors
anchors:
  - [10,13, 16,30, 33,23]  # P3/8
  - [30,61, 62,45, 59,119]  # P4/16
  - [116,90, 156,198, 373,326]  # P5/32

# YOLOv5 backbone
backbone:
  # [from, number, module, args]
  [[-1, 1, Focus, [64, 3]],  # 0-P1/2
   [-1, 1, Conv, [128, 3, 2]],  # 1-P2/4
   [-1, 3, BottleneckCSP, [128]],
   [-1, 1, Conv, [256, 3, 2]],  # 3-P3/8
   [-1, 9, BottleneckCSP, [256]],
   [-1, 1, Conv, [512, 3, 2]],  # 5-P4/16
   [-1, 9, BottleneckCSP, [512]],
   [-1, 1, Conv, [1024, 3, 2]],  # 7-P5/32
   [-1, 1, SPP, [1024, [5, 9, 13]]],
   [-1, 3, BottleneckCSP, [1024, False]],  # 9
  ]

# YOLOv5 head
head:
  [[-1, 1, Conv, [512, 1, 1]],
   [-1, 1, nn.Upsample, [None, 2, 'nearest']],
   [[-1, 6], 1, Concat, [1]],  # cat backbone P4
   [-1, 3, BottleneckCSP, [512, False]],  # 13

   [-1, 1, Conv, [256, 1, 1]],
   [-1, 1, nn.Upsample, [None, 2, 'nearest']],
   [[-1, 4], 1, Concat, [1]],  # cat backbone P3
   [-1, 3, BottleneckCSP, [256, False]],  # 17 (P3/8-small)

   [-1, 1, Conv, [256, 3, 2]],
   [[-1, 14], 1, Concat, [1]],  # cat head P4
   [-1, 3, BottleneckCSP, [512, False]],  # 20 (P4/16-medium)

   [-1, 1, Conv, [512, 3, 2]],
   [[-1, 10], 1, Concat, [1]],  # cat head P5
   [-1, 3, BottleneckCSP, [1024, False]],  # 23 (P5/32-large)

   [[17, 20, 23], 1, Detect, [nc, anchors]],  # Detect(P3, P4, P5)
  ]
````

# Training the model

Let's get familiar with how to run the training script in yolov5 by breaking down the command.

The training command takes in the following options:

- **img:** define input image size
- **batch:** determine batch size
- **epochs:** define the number of training epochs. (Note: often, 3000+ are common here!)
- **data:** set the path to our yaml file
- **project:** the name of our project
- **cfg:** specify our model configuration
- **weights:** specify a custom path to weights. (Note: you can download weights from the Ultralytics Google Drive folder)
- **name:** result names
- **nosave:** only save the final checkpoint
- **cache:** cache images for faster training


We will run the command with:
- **img:** 640
- **batch:** 16
- **epochs:** 20
- **data**: mask-data.yaml (the data yaml file we created earlier)
- **project**: face-mask-detection
- **cfg:** ./models/custom_yolov5s.yaml (the model configuration we just created)
- **device (optional):** 1 (specifies the GPU number to use on your hardware)  <-- You can try removing this if you face any GPU related errors

````python
%cd yolov5
!python train.py --weights yolov5s.pt  --data mask-data.yaml --epochs 20 --batch 16 --img 640 \
--project face-mask-detection --device 1 --cfg ./models/custom_yolov5s.yaml --name yolov5s_results


````
